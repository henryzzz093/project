```python
# import required libraries
import pandas as pd
import numpy as np
```

```python
# read csv file as pd.dataframe
data1 = pd.read_csv('Daily_Demand_Forecasting_Orders-1.csv')
data1
```

```python
# show raw data columns and its dimension
print(data1.columns)
data1.shape
```

```python
# retrieve information from raw data table as a list
A = data1['Week of the month (first week'].values.tolist()

print(A[0])
```

```python
# split data based on ";"
a = []

for i in range(len(A)):
    a.append(A[i].split(';'))
    
print(a[0])
```

```python
# constructing new data table with aligned columns and correct values
new_table = pd.DataFrame(data = a, columns = ['Week of the month', 'Day of the week', 'Non-urgent order', 'Urgent order',
                                      'Order type A', 'Order type B', 'Order type C', 'Fiscal sector orders', 
                                      'Orders from the traffic controller sector', 'Banking orders (1)',
                                       'Banking orders (2)', 'Banking orders (3)', 'Target (Total orders)'])
new_table
```

```python
# convert entire dataframe to numerical value
print(new_table.dtypes)
new_table = new_table.apply(pd.to_numeric)
print(new_table.dtypes)

# make a copy of complete table for reference:
complete_table = new_table
# complete_table
```

```python
# basic five statistical summary about our data
new_table.describe()
```

```python
# show data distributions
%matplotlib inline
import matplotlib.pyplot as plt
new_table.hist(bins = 20, figsize = (20,15))
plt.show()
print("Banking Order 1,2,3 and Orders from the traffic controller sector are using differnt scales with other attributes")
```

```python
# create test set (20% of the dataset)
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(new_table, test_size = 0.2, shuffle = False)
print(len(train_set))
print(len(test_set))
```

```python
# check for missing values
new_table.isnull().values.any()
```

```python
# Trying out various attribute combinations

# combine non-urgent order with urgent order:
new_table['Non_ur_+_Urg'] = new_table['Non-urgent order'] + new_table['Urgent order']

# combine all types (A+B+C) order:
new_table['all_types_order'] = new_table['Order type A'] + new_table['Order type B'] + new_table['Order type C']

# combine all banking orders:
new_table['all_banking_order'] = new_table['Banking orders (1)']+ new_table['Banking orders (2)'] + new_table['Banking orders (3)']

# combine all sector orders:
new_table['all_sector_order'] = new_table['Fiscal sector orders'] + new_table['Orders from the traffic controller sector']
new_table
                                                                                                    
```

```python
# looking for correlations between each columns 
corr_matrix = new_table.corr()
corr_matrix
```

```python
# correlations to target column ['Target (Total orders)']
print(corr_matrix['Target (Total orders)'].sort_values(ascending = False))
print("****************************************************")
print("Based on the correlation, we can see that there are numbers of features shows strong correlation to our target column") 
print("let us try linear regression modeling for now")
print("from the corr matrix, we can see that target = all_types_order")
print("let's use Non_ur+Urg be the features to train our modeling")
```

```python
# initiate first regression modeling
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# select training and testing data
X_train = train_set[['Urgent order', 'Non-urgent order']]
y_train = train_set['Target (Total orders)']

X_test = test_set[['Urgent order', 'Non-urgent order']]

# training process
regression = LinearRegression()
regression.fit(X_train, y_train)

# get predict values
predictedTotal = regression.predict(X_test)
print(predictedTotal)
answer = test_set['Target (Total orders)'].values
print(answer)

# check for performance:
lin_mse = mean_squared_error(predictedTotal, answer)
lin_rmse = np.sqrt(lin_mse)

print(lin_rmse)
print("*************************************************************")
print("The mean of Total Order is 300.873317, so the average prediction error of {} is not that great".format(lin_rmse))
print("Considering add more features: Order type B,C")
```

```python
# Add two more features into our training set, and see how it goes
X_train_2 = train_set[['Urgent order', 'Non-urgent order', 'Order type B', 'Order type C']]
y_train_2 = train_set['Target (Total orders)']

X_test_2 =  test_set[['Urgent order', 'Non-urgent order', 'Order type B', 'Order type C']]

# training process
regression_2 = LinearRegression()
regression_2.fit(X_train_2, y_train_2)

# get predict values
predictedTotal_2 = regression_2.predict(X_test_2)
print(predictedTotal_2)
answer = test_set['Target (Total orders)'].values
print(answer)

# check for performance:
lin_mse_2 = mean_squared_error(predictedTotal_2, answer)
lin_rmse_2 = np.sqrt(lin_mse_2)

print(lin_rmse_2)
print("*************************************************************")
print("with the Average prediction error decrease to {}, the model now seems better".format(lin_rmse_2))
```

```python
# increasing model complexity, try different modeling method ---- Decision Tree
from sklearn.tree import DecisionTreeRegressor

# training model
tree_reg = DecisionTreeRegressor(random_state = 42)
tree_reg.fit(X_train_2, y_train_2)
tree_pred = tree_reg.predict(X_test_2)

# get prediction errors
tree_mse = mean_squared_error(tree_pred, answer)
tree_rmse = np.sqrt(tree_mse)
tree_rmse

print("DecisionTree seems like not a good pick for our model since the root mean squared error has increase to {}".format(tree_rmse))
```
